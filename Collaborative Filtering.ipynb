```python
import numpy as np
import pandas as pd
```

```python
# Read in the datasets
movies = pd.read_csv('data/movies.csv')
reviews = pd.read_csv('data/ratings.csv')

reviews.info()

# using a subset of the data
reviews = reviews[reviews['user_id'] <= 15_000]
```

```python
user_items = reviews.pivot_table(index='user_id', columns='movie_id', values='rating', fill_value=0)
user_items = user_items.astype('int8')

user_items.head()
```

```python
def moviesWatced(user_id,user_items):

    movies = user_items.columns[user_items.loc[user_id]>0].tolist()

    return movies
```

```python
def createUserMovieDict(user_items):

    moviesSeen = dict()

    for user in user_items.index:

        moviesSeen[user] = moviesWatced(user,user_items)


    return moviesSeen
        
moviesSeen = createUserMovieDict(user_items)
```

```python
# clean the data, keep users with at least 3 recommendations

def cleanUserMovieDict(moviesSeen, minimum = 2 ):

    moviesToUse = {}

    for user, movies in moviesSeen.items():
        if len(movies) > minimum:
            moviesToUse[user] = movies

    return moviesToUse

moviesToAnalyze = cleanUserMovieDict(moviesSeen)
```

#### Pearson's Correlation Coefficient 

```python
def pearson_corr(x, y):


    # Compute Mean Values
    mean_x, mean_y = np.mean(x), np.mean(y) 
    
    # Compute diffs
    x_diffs = x - mean_x
    y_diffs = y - mean_y

    numerator = np.sum(x_diffs * y_diffs)
    denominator = np.sqrt(np.sum(x_diffs**2)) * np.sqrt(np.sum(y_diffs**2))
        
    corr = numerator/denominator
                            
    return corr 
```
